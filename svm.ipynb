```python
X = df.iloc[:, 1:22]
X = X.join(df.iloc[:, 24:25])

X
```

```python
# label encoding  X
from sklearn import preprocessing

label_encoder = preprocessing.LabelEncoder()

for column in X:
  X[column] = label_encoder.fit_transform(X[column])

X
```

```python
# standardization
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
X_cont_SS = SS.fit_transform(X)
X_cont_SS = pd.DataFrame(X_cont_SS)
X_cont_SS.columns = list(X)
X_cont_SS.head()
```

```python
# data partition
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X_cont_SS,Y,test_size=0.30)
```

```python
#model fitting

from sklearn.svm import SVC

model= SVC(kernel='rbf',gamma='scale')

model.fit(X_train.values,Y_train)
Y_pred_train = model.predict(X_train.values)
Y_pred_test = model.predict(X_test.values)
```

```python
model = SVC(C = 2,kernel='linear')
model.fit(X_train.values,Y_train)
Y_pred_train = model.predict(X_train.values)
Y_pred_test = model.predict(X_test.values)
```

```python
model = SVC(kernel='poly',degree=3)

model.fit(X_train.values,Y_train)
Y_pred_train = model.predict(X_train.values)
Y_pred_test = model.predict(X_test.values)
```

```python
# Strengths of SVM for the Mushroom dataset:

# 1. High accuracy:
# SVM achieved high accuracy on both training and test sets, indicating its effectiveness in classifying edible and poisonous mushrooms.

# 2. Robustness to noise and outliers:
# SVM is robust to noise and outliers due to its maximum margin principle, which helps in separating the classes even in the presence of noisy data points.

# 3. Effective with non-linearly separable data:
# SVM can be used with different kernel functions to handle non-linearly separable data, making it suitable for the Mushroom dataset, which may have complex relationships between features and class labels.

# 4. Interpretability:
# SVM provides interpretable models, allowing users to understand the decision boundaries and the features contributing to the classification.



# Weaknesses of SVM for the Mushroom dataset:

# 1. Computational cost:
# Training an SVM model can be computationally expensive, especially for large datasets.

# 2. Parameter tuning:
# SVM has several parameters that need to be tuned for optimal performance, which can be time-consuming and requires expertise.

# 3. Sensitivity to imbalanced data:
# SVM may not perform well on imbalanced datasets where one class is significantly larger than the other.

# 4. Limited ability to handle missing data:
# SVM does not inherently handle missing data, and preprocessing techniques need to be applied to address missing values.

# 5. Potential overfitting:
# SVM models with high complexity can overfit the training data, leading to poor generalization performance on unseen data.
```
